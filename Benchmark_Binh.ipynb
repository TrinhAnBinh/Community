```python
import requests
import json
import pandas as pd
import datetime as dt
import pandas as pd
import numpy as np
import sys

# import matplotlib.pyplot as plt
# import plotly.graph_objects as go
# import plotly.express as px
```

```python
def getTradingData(fromDate, toDate, ticker):
    """
    Khởi tạo chuỗi thời gian giao dịch thực tế.
    
    Args:
        **fromDate** (datetime): Ngày bắt đầu \n
        **toDate** (datetime): Ngày kết thúc
        **ticker** (string): Tên ticker
    
    Returns:
        pandas.DataFrame.
    """
    
    fDate = int(fromDate.timestamp())
    tDate = int(toDate.timestamp())
    data = requests.get(f"https://api.vietstock.vn/ta/history?symbol={ticker}&resolution=D&from={fDate}&to={tDate}").json()
    data2 = json.loads(data)
    df0 = pd.DataFrame(data2)
#     df0["t"] = pd.to_datetime(df0["t"],unit="s").apply(lambda x: dt.datetime(x.year,x.month,x.day))
    df0["t"] = pd.to_datetime(df0["t"],unit="s")
    df0["symbol"]=ticker
    df0.columns=['TradingDate','Open','High','Low','Close','Volume','Status','Ticker']
    return df0
```

```python
# fDate = dt.datetime(2019,6,28)
# tDate = dt.datetime(2019,7,23)
tickers=["VNINDEX", "HNXINDEX", "UPCOMINDEX"]

data_2000_2020 = {}
for i in tickers:
#     data_2000_2005=getTradingData(dt.datetime(2000,1,2),dt.datetime(2005,12,31) , i)
#     data_2006_2010=getTradingData(dt.datetime(2006,1,2),dt.datetime(2010,12,31) , i)
#     data_2011_2015=getTradingData(dt.datetime(2011,1,2),dt.datetime(2015,12,31) , i)
    data_2016_2019=getTradingData(dt.datetime(2020,3,15),dt.datetime(2020,4,24) , i)
    data_2000_2020[i] = pd.concat([data_2016_2019])  

totalData = pd.concat([data_2000_2020[i] for i in tickers])
totalData
```

```python
totalData['day_of_week'] = totalData['TradingDate'].dt.day_name()
Data_model=totalData.loc[(totalData["day_of_week"]=="Friday")]
```

```python
Data_model_temp=Data_model[["TradingDate","Ticker","Close","day_of_week"]]
# Data_model_temp["Return"]=Data_model_temp["Close"].pct_change()
Data_model_temp
```

```python
# newData_temp=Data_model_temp.pivot_table(columns='Ticker', index="TradingDate", aggfunc="first").stack("Ticker").unstack().bfill(axis='rows')
newData_temp=Data_model_temp.pivot_table(columns='Ticker', index="TradingDate", aggfunc="first").stack("Ticker")
```

```python
# newData_temp
ChartData=newData_temp.reset_index()
ChartData
```

```python
def caculate_return(df_name):
    new_df = ChartData[ChartData["Ticker"] == df_name]
    new_df["return"] = new_df["Close"]/float(new_df["Close"][new_df["TradingDate"]==ChartData["TradingDate"].min()])-1
    return new_df

VNINDEX =caculate_return("VNINDEX")
UPCOMINDEX = caculate_return("UPCOMINDEX")
HNXINDEX = caculate_return("HNXINDEX")
frames = [VNINDEX,HNXINDEX,UPCOMINDEX]
result = pd.concat(frames)
result
```

```python
new_data_chart=result.pivot_table(columns='Ticker', index="TradingDate", aggfunc="first").stack("Ticker").unstack().bfill(axis='rows')
new_data_chart
```
